# EDA

# Distribution of Room types

```python
room_counts = df["room_type"].value_counts()
plt.figure(figsize=(3,3))
shuffled = room_counts.sample(frac=1, random_state=42) 
plt.pie(
    shuffled, 
    labels=room_counts.index, 
    autopct='%1.1f%%',       
    startangle=0,          
    colors = ["#FFB6B9", "#FAE3D9", "#BBDED6", "#8AC6D1"] 
)
plt.title("Room Type Distribution")
plt.axis("equal")  
plt.show()
```

```python
df = df[df['price'] <1500]
```

# Distribution of the prices 

```python
sns.histplot(data = df , x ='price' ,color = 'lightgreen',bins = 100)
plt.title("Distribution of the prices")
```

```python
df.columns
```

 # Distribution of number rooms available for booking corresponds to the number of days available out of 365 days.

```python
plt.figure(figsize=(6,3))
sns.histplot(data =df,x = 'availability_365',bins = 20, color = "orange", hue="room_type", multiple="stack",)
plt.title('Availability 365 distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby(by ='neighbourhood_group')['price'].mean() 
```

# Dependency of price on neighbourhood groups

```python
sns.barplot(data = df,x ='neighbourhood_group', y ='price',hue = 'room_type')
```

# Number of Reviews and Price Relationship 

```python
plt.figure(figsize =(7,5))
sns.scatterplot(data =df, x ='number_of_reviews',y = 'price', hue ='room_type')
plt.title("No. of reviews and price relationship")
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue ='room_type')
```

```python
sns.scatterplot(data = df ,x='longitude',y='latitude', hue ="room_type")
plt.title("Geographical distribution of Airbnb Listing")
```

```python
co = df[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
co
plt.figure(figsize=(8,6))
sns.heatmap(data =co,annot=True)
```
